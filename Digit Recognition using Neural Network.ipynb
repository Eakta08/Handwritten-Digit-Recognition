## Importing libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

## Reading dataset

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
X_train
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
plt.imshow(X_train[0])
```

```python
y_train[0]
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
plt.imshow(X_train[0])
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened[0]
```

```python
X_train_flattened.shape
```

## Model

* ### Neural Network using no hidden layer

In this Neural Network model
- Input layers contains 784 input nodes 
- Output layer contains 10 output nodes for classification of target labels (uses activation function - sigmoid)

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train_flattened, y_train, epochs=10)
```

```python
model.summary()
```

```python
metric=model.evaluate(X_test_flattened,y_test)
metric
```

```python
print("loss ---> ",metric[0])
print("accuracy --->",metric[1])
```

```python
y_pred = model.predict(X_test_flattened)
y_pred
```

```python
y_pred[1]
```

```python
np.argmax(y_pred[1])
```

```python
plt.imshow(X_test[1])
```

```python
y_test[1]
```

```python
y_pred_label=[np.argmax(i) for i in y_pred]
y_pred_label[:10]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_label)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

* ### Neural Network using a hidden layer

In this Neural Network model
- Input layer contains 784 input nodes
- Hidden layer contains 100 hidden layer nodes (uses activation function - relu)
- Output layer contains 10 output nodes for classification of target labels (uses activation function - sigmoid)

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train_flattened, y_train, epochs=10)
```

```python
model.summary()
```

```python
metric=model.evaluate(X_test_flattened,y_test)
metric
```

```python
print("loss ---> ",metric[0])
print("accuracy --->",metric[1])
```

```python
y_pred = model.predict(X_test_flattened)
y_pred
```

```python
y_pred[1]
```

```python
np.argmax(y_pred[1])
```

```python
plt.imshow(X_test[1])
```

```python
y_test[1]
```

```python
y_pred_label=[np.argmax(i) for i in y_pred]
y_pred_label[:10]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_label)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

* ### Neural Network using 2 hidden layer

In this Neural Network model
- Input layer contains 784 input nodes
- 1st Hidden layer contains 200 hidden layer nodes (uses activation function - sigmoid)
- 2nd Hidden layer contains 100 hidden layer nodes (uses activation function - sigmoid)
- Output layer contains 10 output nodes for classification of target labels (uses activation function - sigmoid)

```python
model=keras.Sequential([
    keras.layers.Dense(200,input_shape=(784,), activation='sigmoid'),
    keras.layers.Dense(100, activation='sigmoid'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train_flattened, y_train, epochs=10)
```

```python
model.summary()
```

```python
metric=model.evaluate(X_test_flattened,y_test)
metric
```

```python
print("loss ---> ",metric[0])
print("accuracy --->",metric[1])
```

```python
y_pred = model.predict(X_test_flattened)
y_pred
```

```python
y_pred[1]
```

```python
np.argmax(y_pred[1])
```

```python
plt.imshow(X_test[1])
```

```python
y_test[1]
```

```python
y_pred_label=[np.argmax(i) for i in y_pred]
y_pred_label[:10]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_label)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

* ### Neural Network Using Flatten layer so that we don't have to call .reshape on input dataset

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(200, activation='sigmoid'),
    keras.layers.Dense(100, activation='sigmoid'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```
