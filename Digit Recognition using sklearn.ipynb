# Handwritten Digit Recognition

_This jupyter notebook contains digit recognition using sklearn library using Logistic Regression, KNN, SVM, Decision Tree and Random Forest. Hope you enjoy it..._

## Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Print to show there are 1797 images (8 by 8 images for a dimensionality of 64)
print("Image Data Shape" , digits.data.shape)

# Print to show there are 1797 labels (integers from 0-9)
print("Label Data Shape", digits.target.shape)
```

## Data Visualization

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize = 20)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=88)
```

## Model Selection

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.model_selection import cross_val_score
```

```python
com=pd.DataFrame({'Model':[],'Training Accuracy Score':[],'Testing Accuracy Score':[],'Cross Validation Score':[]})
com
```

* ### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test[0].reshape(1,-1))
```

```python
lr.predict(X_test[0:10])
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
y_train_pred=lr.predict(X_train)
print("Training Accuracy Score: ",accuracy_score(y_train,y_train_pred))
```

```python
print("Testing Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
print("Cross Validation Score:",cross_val_score(lr,X_train,y_train,cv=10).mean())
```

```python
com.loc[1]=["Logistic Regression", accuracy_score(y_train,y_train_pred), accuracy_score(y_test,y_pred), cross_val_score(lr,X_train,y_train,cv=10).mean()]
com
```

```python
cm = confusion_matrix(y_test,y_pred)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(accuracy_score(y_test,y_pred))
plt.title(all_sample_title, size = 15);
```
